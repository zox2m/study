```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
font_path= "malgun.ttf"
font_name=font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)

#excel   
df = pd.read_excel('  .xlsx',header=0)

#NAN    (   )
df = df.fillna(method='ffill')

#      
mask = (df[''] == '') & (df[''] != '')
df_seoul= df[mask]
df_seoul= df_seoul.drop([''],axis=1)
df_seoul.rename({'':''},axis=1,inplace=True)
df_seoul.set_index('',inplace=True)

#      
sr_one=df_seoul.loc['']

#x y  plot 
plt.plot(sr_one.index,sr_one.values)
#  plot  
plt.plot(sr_one)

#      
sr_one = df_seoul.loc['']

#x,y  plot  
plt.plot(sr_one.index,sr_one.values)

#  
plt.title(' ->   ')

#  
plt.xlabel('')
plt.ylabel(' ')

plt.show()
```

```python
#  
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

#axe  plot   
ax1.plot(sr_one,'o',markersize=10)
ax2.plot(sr_one,marker='o',markerfacecolor='green',markersize=10,
        color='olive',linewidth=2,label=' -> ')
ax2.legend(loc='best')

#y   (  )
ax1.set_ylim(50000,800000)
ax2.set_ylim(50000,800000)

#     75 
ax1.set_xticklabels(sr_one.index,rotation=75)
ax2.set_xticklabels(sr_one.index,rotation=75)

plt.show()
```

```python
#  
fig=plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

# axe  plot   
ax.plot(sr_one,marker='o',markerfacecolor='orange',markersize=10,color='olive',linewidth=2,label=' -> ')
ax.legend(loc='best')

#y  
ax.set_ylim(50000,800000)

#  
ax.set_title(' ->   ',size=20)

#  
ax.set_xlabel('',size=12)
ax.set_ylabel('',size = 12)

#     75 
ax.set_xticklabels(sr_one.index,rotation=75)

#   
ax.tick_params(axis='x',labelsize=10)
ax.tick_params(axis='y',labelsize=10)

plt.show()
```

```python
# . .      
col_years = list(map(str,range(1970,2018)))
df_3 = df_seoul.loc[['','',''],col_years]

#  
plt.style.use('ggplot')

#  
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)

#axe  plot   
ax.plot(col_years,df_3.loc['',:],
        marker='o',markerfacecolor='green',markersize=10,
        color='olive',linewidth=2,label=' -> ')
ax.plot(col_years,df_3.loc['',:],
        marker='o',markerfacecolor='blue',markersize=10,
        color='skyblue',linewidth=2,label=' -> ')
ax.plot(col_years,df_3.loc['',:],
        marker='o',markerfacecolor='red',markersize=10,
        color='magenta',linewidth=2,label=' -> ')

# 
ax.legend(loc='best')

#  
ax.set_title(' -> , , ',size=20)

#  
ax.set_xlabel('',size=12)
ax.set_ylabel(' ',size=12)

#     90 
ax.set_xticklabels(col_years,rotation=90)

#   
ax.tick_params(axis='x',labelsize=10)
ax.tick_params(axis='y',labelsize=10)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc

font_path = "malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

plt.style.use('ggplot')
plt.rcParams['axes.unicode_minus']=False

#excel    
df=pd.read_excel('.xlsx',convert_float=True)
df=df.loc[5:9]
df.drop(' (h)',axis='columns',inplace=True)
df.set_index(' ',inplace=True)
df=df.T

# 
df = df.rename(columns = {'': ''})
df[' - 1'] = df[''].shift(1)
df[''] = ((df['']/df[' - 1'])-1)*100

#2  
ax1=df[['','']].plot(kind='bar',figsize=(20,10),width=0.7,
                        stacked=True)
ax2=ax1.twinx()
ax2.plot(df.index,df.,ls='',marker='o',markersize=20,
        color='red',label=' (%)')

ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)

ax1.set_xlabel('',size=20)
ax2.set_ylabel('( Kwh)')
ax2.set_ylabel('   (%)')

plt.title('   (1990~ 2016)',size=10)

ax.legend(loc='upper left')

plt.show()
```
