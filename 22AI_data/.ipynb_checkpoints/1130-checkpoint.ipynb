```python
import matplotlib.pyplot as plt
import seaborn as sns

#seaborn 제공 데이터셋 가져오기
titanic = sns.load_dataset('titanic')

#스타일 테마 설정
sns.set_style('whitegrid')

#그래프 객체 생성(figure에 4개의 서브 플롯 생성)
fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

#박스 플롯 
sns.boxplot(x='alive',y='age',data=titanic,ax=ax1)
            
#박스 그래프 - hue 추가
sns.boxplot(x='alive',y='age',hue='sex',data=titanic,ax=ax2)

#바이올린 플롯
sns.violinplot(x='alive',y='age',data=titanic,ax=ax3)

#바이올린 플롯 - hue
sns.violinplot(x='alive',y='age',hue='sex',data=titanic,ax=ax4)

plt.show()
            
            
```

```python
#조인트 그래프 
```

```python
#조인트 그래프 기본값( 산점도)
j1 = sns.jointplot(x='fare',y='age',data=titanic)

#조인트 - 회귀선
j2 = sns.jointplot(x='fare',y='age',kind='reg',data=titanic)
#조인트 - 육각 그래프
j3 = sns.jointplot(x='fare',y='age',kind='hex',data=titanic)
#조인트 - 커럴 밀집 그래프 
j4 = sns.jointplot(x='fare',y='age',kind='kde',data=titanic)


#챠트 제목 표시
j1.fig.suptitle('titanic fare - scatter',size=15)
j2.fig.suptitle('titanic fare - reg',size=15)
j3.fig.suptitle('titanic fare - hex',size=15)
j4.fig.suptitle('titanic fare - kde',size=15)

plt.show()
```

```python
g = sns.FacetGrid(data=titanic,col='who',row='survived')


#그래프 적용하기
g = g.map(plt.hist,'age')
```

```python
#titanic 데이터셋 중에서 분석 데이터 선택하기
titanic_pair = titanic[['age','pclass','fare']]

#조건에 따라 그리드 나누기
g=sns.pairplot(titanic_pair)

```

```python
import pandas as pd
df = pd.read_excel('서울지역 대학교 위치.xlsx')

seoul_map = folium.Map(location =[37.55,126.98],ittles='Stamen Terrain',
                        zoom_start=12)

for name, lat,lng in zip(df.index, df.위도, df.경도):
    folium.Marker([lat,lng],popup=name).add_to(seoul_map)
    
    #지도를 HTML  파일로 저장하기
    seoul_map.save('seoul_colleges.html')
    
    
```

```python
for name, lat,lng in zip(df.index, df.위도, df.경도):
    folium.CircleMarker([lat,lng],
                        radius=10,
                        color='brown',
                        fill=True,
                        fill_color='coral',
                        fill_opacity=0.7,
                        popup=name).add_to(seoul_map)
    seoul_map.save('seoul_colleges2.html')
```

```python
import json

file_path = '경기도인구데이터.xlsx'
df = pd.read_excel(file_path,index_col='구분')
df.columns = df.columns.map(str)

#경기도 시군구 경계 정보를 가진 geo-json 파일 불러오기
geo_path = '경기도행정구역경계.json'
try:
    geo_data = json.load(open(geo_path,encoding='utf-8'))
except:
    geo_data = json.load(open(geo_path,encoding='utf-8-8ig'))
    
#경기도 지도 만들기
g_map = folium.Map(location =[37.55,126.98],ittles='Stamen Terrain',
                        zoom_start=9)
#출력할 연도 선책
year = '2007'

#choropleth 클래스로 단계구분도 표시하기

folium.Choropleth(geo_data=geo_data,
                 data=df[year],
                 columns = [df.index,df[year]],
                 dill_color ='YlorRd',fill_opacity=0.7, line_opacity=0.3,
                 threshold_scale=[10000,10000,30000,500000,700000],
                 key_on='feature.properties.name')
                #지도를 파일로 저장
g_map.save('gyonggi_population_'+ year+'.html')
```
